```python
Job_data = {
    "Job_ID": [1, 2, 3, 4, 5],
    "Role":["Data Analyst", "Datat Scientist", "Business Analys", "Data Engineer", "ML Engineer"],
    "Company":["ABC Corp", "Xyz Ltd", "Data works", "Infratech", "AI Labs"],
    "Skills":[
        "Excel SQL Python",
        "Python ML Statistics",
        "Excel SQL PwerBI",
        "Python SQL Spark",
        "Python ML DeepLearning"

    ]
}
for i in range(len(Job_data["Job_ID"])): # Corrected 'Job_data' to 'Job_data'
  print(f"ID: {Job_data['Job_ID'][i]} | Role:{Job_data['Role'][i]} | Company:{Job_data['Company'][i]} | Skills:{Job_data['Skills'][i]}")
```

```python
n = int(input("How many job posting do you want to add: "))

for i in range(n):
  role = input("Enter role: ")
  company = input("Enter company: ")
  skills = input("Enter skills: ")

  new_id = max(Job_data["Job_ID"]) + 1

  Job_data["Job_ID"].append(new_id)
  Job_data["Role"].append(role)
  Job_data["Company"].append(company)
  Job_data["Skills"].append(skills)

print("Job postings updated successfully!")
print(Job_data)
```

```python
def clean_skills(skill_text):
  skill_text = skill_text.lower()
  skill_text = skill_text.replace(","," ")
  skill_text = "".join(skill_text.split())
  return skill_text

Job_data["Skills"] = [clean_skills(s) for s in Job_data["Skills"]]
for skill in Job_data["Skills"]:
  print(skill)
```

```python
def find_jobs_by_skill(skill_query):
  skill_query = skill_query.lower().strip()
  results = []
  for i in range(len(Job_data["Skills"])):
    skills = Job_data["Skills"][i].split()
    if skill_query in skills:
      results.append(Job_data["Job_ID"][i]) # Corrected 'Job_Id' to 'Job_ID'
      results.append(Job_data["Role"][i])
      results.append(Job_data["Company"][i])
  if results :
    return results
  else:
    return "No jobs found"
```
